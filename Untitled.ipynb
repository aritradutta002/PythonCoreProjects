### Problem 1:

1. Write a program to calculate area of shapes. Your program should be capable of calculating the area of a square, rectangle, triangle and a circle. The user should be presented with options to select the shape. Based on which shape is chosed by the user, the program should ask for the appropriate input and print the resulting area on the screen.

When the program is run, the screen should display something like this:

**Which shape would you like to calculate the area for? Please enter the option number-**<br>
**1. Square**<br>
**2. Rectangle**<br>
**3. Triangle**<br>
**4. Circle**<br>
**Enter Option: _**

Say the user enters the option 1.

**Please enter the length of a side: _**

If the user enters a value of 5. The output should be:

**The area of the square is 25**

This program should indicate that the input is invalid if the user enters a character instead of a number as input. For instance if the user enters a value of ‘a’ instead of 5 in the previous example the program should prompt:

**Invalid input, please enter a number: _**


```python
def square():
    while True:
        try:
            n = float(input('Please enter the length of a side:'))
            if n <= 0:
                raise OSError('Invalid input, please enter a positive number!!')
            print('The area of the square is',n**2)
            break
        except OSError as e:
            print(e)
        except:
            print('Invalid input, please enter a number!!')
def rectangle():
    while True:
        try:
            x = float(input('Please enter the length of the rectangle:'))
            y = float(input('Please enter the width of the rectangle:'))
            if x <= 0 or y<= 0:
                raise OSError('Invalid input, please enter a positive number!!')
            print('The area of the rectangle is',x*y)
            break
        except OSError as e:
            print(e)
        except:
            print('Invalid input, please enter a number!!')
def triangle():
    while True:
        try:
            a = float(input('Please Enter a side of the triangle:'))
            b = float(input('Please Enter another side of the triangle:'))
            c = float(input('Please Enter another side of the triangle:'))
            if a <= 0 or b <= 0 or c <= 0:
                raise OSError('Invalid input, please enter a positive number!!')
            s = (a+b+c)/2
            area = (s*(s-a)*(s-b)*(s-c))**0.5
            print('The area of the triangle is',area)
            break
        except OSError as e:
            print(e)
        except:
            print('Invalid input, please enter a number!!')
def circle():
    while True:
        try:
            pi = 22/7
            r = float(input('Please enter the radius of the circle:'))
            if r <= 0:
                raise OSError('Invalid input, please enter a positive number!!')
            print('The area of the circle is',pi*r*r)
            break
        except OSError as e:
            print(e)
        except:
            print('Invalid input, please enter a number!!')
def switch_case(c):
    switcher = {1:square,2:rectangle,3:triangle,4:circle}
    func = switcher.get(c,lambda:"Invalid input, please enter a number:")
    func()
while True:
    try:
        choice = int(input('Which shape would you like to calculate the area for? Please enter the option number-\n1. Square\n2. Rectangle\n3. Triangle\n4. Circle\nEnter Option:'))
        if choice not in [1,2,3,4]:
            print("\n!!Invalid input, please enter a number in range!!\n")
            continue
        switch_case(int(choice))
        break
    except:
        print('Invalid input, please enter a number!!')
```
